# 1. Линейная алгебра

1. Даны матрицы:<br>

$A = \begin{pmatrix}
  -2 & -4\\
  3 & 5\\
  -1 & 0
 \end{pmatrix}$
 $B = \begin{pmatrix}
  1 & 2 & 7\\
  -3 & -4 & 0\\
  5 & 2 & 1
 \end{pmatrix}$
 $C = \begin{pmatrix}
  6 & -3 & 9\\
  4 & -5 & 2\\
  8 & 1 & 5
 \end{pmatrix}$
<br>
<br>
Посчитайте матрицу $D=A^⊤C−2A^⊤B^⊤$.
<br>
Приведите полную последовательность вычислений.

```python
# Проверка
import numpy as np

A = np.array([[2,-4],[3,5],[-1,0]])
B = np.array([[1,2,7],[-3,-4,0],[5,2,1]])
C = np.array([[6,-3,9],[4,-5,2],[8,1,5]])
D = np.dot(A.T, C) - np.dot(2*A.T, B.T)

print(D)
```
