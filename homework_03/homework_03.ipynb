## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def converter(lc_str: str) -> str:
    if lc_str[0].isupper():
        result_str = lc_str[0].lower()
        for char in lc_str[1:]:
            if char.isupper():
                result_str += '_'
                result_str += char.lower()
            else:
                result_str += char
        return result_str

    elif lc_str.find('_'):
        lst = lc_str.split('_')
        return ''.join(map(lambda word: word.capitalize(), lst))

    else:
        return lc_str

input_str = input('Enter string')
print(input_str, ' -> ', converter(input_str))
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def is_leap_year(year: int) -> bool:
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            else:
                return False
        else:
            return False
    else:
        return False

def is_a_date(input_str: str) -> bool:
    dct_days = {1:31, 2:29, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}

    parts_of_the_date = input_str.split('.')
    parts_of_the_date = list(map(int, parts_of_the_date))
    if (parts_of_the_date[2] < 1 # проверка года, месяца и дня

        or parts_of_the_date[1] > 12
        or parts_of_the_date[1] < 1

        or (parts_of_the_date[0] < 1
        or parts_of_the_date[0] > dct_days[parts_of_the_date[1]])):

        return False

    elif parts_of_the_date[1] == 2: # проверка февраля
        is_leap = is_leap_year(parts_of_the_date[2])
        if not is_leap and parts_of_the_date[0] == 29 :
            return False
        else:
            return True
    else:
        return True

print('29.16.2000', ' -> ', is_a_date('29.16.2000'))
print('29.02.2000', ' -> ', is_a_date('29.02.2000'))
print('29.02.2001', ' -> ', is_a_date('29.02.2001'))
print('31.04.1962', ' -> ', is_a_date('31.04.1962'))
print('22.03.1983', ' -> ', is_a_date('22.03.1983'))
print('29.02.1900', ' -> ', is_a_date('29.02.1900'))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_simple(number:int) -> bool:
    for i in range(2, number):
        if number % i == 0:
            return False

    return True


print(17, ' -> ', is_simple(17))
print(20, ' -> ', is_simple(20))
print(23, ' -> ', is_simple(23))
```
