## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
input_number = input("Input number: ")
if input_number.isdecimal():
    print(input_number[0] + input_number[3:0:-1] + input_number[4])

```

## Задание 2: Подсчет выходных до отпуска

* `14 -> 4`

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
parameters = input("Input length, width and the size: ")
parameters = parameters.split(",")

length = int(parameters[0].rstrip())
width = int(parameters[1].rstrip())
size = int(parameters[2].rstrip())

if size%length == 0 or size%width == 0:
    print("True")
else:
    print("False")

```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
""""
I = 1
V = 5
X = 10
L = 50
C = 100
D = 500
M = 1000"""

input_number = input("Input number: ")
if input_number.isdecimal():
    number = int(input_number)
    if number < 4000:
        result_number = ""
        sample = "MDCLXVI"
        sample_number = 1000

        for i in range(7):
            count_for_letter = int(number // sample_number)
            number -= count_for_letter * sample_number
            result_number += sample[i] * count_for_letter

            if number != 0:
                if int(str(number)[0]) == int(str(sample_number)[0]) - 1:
                    result_number += sample[i+1]
                    result_number += sample[i]
                    number = number - int(str(number)[0] + str(sample_number)[1:])

            if str(sample_number)[0] == "1":
                sample_number //= 2
            else:
                sample_number //= 5

        print(result_number)
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
input_number = input("Input number: ")
str_parts = input_number.split(".")

result = True
if input_number.count(".") > 1:
    result = False

for part in str_parts:
    part = part.strip()
    if part != "" and not part.isdecimal():
        result = False

print(result)
```
